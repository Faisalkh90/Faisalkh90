# A*
A* (A star) algorithm is a search algorithm that finds the shortest path from the starting node to the goal node. It combines two algorithms: **greedy search algorithm** and **uniform cost search algorithm** to form one reliable algorithm.A* is considered an **informed search algorithm**. informed search algorithms are implemented using something called **“heuristic function”**. Heuristic function is like a prediction mechanism to predict the cost of an ongoing edge , so we don't have to reach unnecessary nodes.As shown in the upper figure we have a **heuristic table** which gives us the heuristic of each node, but as you can see the heuristic does not equal the actual cost which is normal because its a prediction it does have to be 100%. the heuristic of node B is 4 and the heuristic of node A  is 12 so the algorithm will chose node B and so on . 

- g(n) — this represents the exact cost of the path from the starting node to any node n.
- h(n) — this represents the heuristic estimated cost from node n to the goal node.
- f(n) — lowest cost in the neighboring node n ,f(n)=g(n)+h(n).



# A*
خوارزمية A * هي خوارزمية بحث تعثر على أقصر مسار من عقدة البداية إلى عقدة الهدف. فهي تجمع بين خوارزميتين: خوارزمية البحث الجشع(greedy search)ايضا مع خوارزمية البحث عن التكلفة الموحدة(uniform cost search) لتشكيل خوارزمية واحدة موثوقة. تعتبر A * خوارزمية بحث مستنيرة(informed algorithm). يتم تنفيذ خوارزميات البحث مستنيرة باستخدام شيء يسمى "الوظيفة الكشف عن مجريات الأمور" (heuristic function). تشبه آلية التنبؤ بتكلفة الحافة المستمرة ، لذلك لا يجيب علينا الوصول إلى عقد غير ضرورية. كما هو موضح في الشكل العلوي ، لدينا جدول إرشادي يعطينا الدليل الإرشادي لكل عقدة ، ولكن كما هو واضح الاستدلال لا يساوي التكلفة الفعلية وهو أمر طبيعي لأنه توقع لا يجب أن يكون 100٪. إرشاد العقدة B هو 4 وإرشاد العقدة A هو 12 ، لذا ستختار الخوارزمية العقدة B وما إلى ذلك.

- g (n) - يمثل هذا التكلفة الدقيقة للمسار من عقدة البداية إلى أي عقدة n.
- h (n) - يمثل هذا التكلفة التقديرية الإرشادية من العقدة n إلى العقدة الهدف.
- f (n) - أقل تكلفة في العقدة المجاورة  ، f (n) = g (n) + h (n).

# A* Algorithm in pseudocode

```python
def a_star_search(tree, heuristic):
    cost = {start: 0}
    ExpandedNodes = [] 
    Fringe        = [[start, heuristic[start]]] 


    while True:
        fn = [i[1] for i in Fringe]
        chosen_index = fn.index(min(fn))
        node = Fringe[chosen_index][0]
        ExpandedNodes.append(Fringe[chosen_index]) 
        del Fringe[chosen_index]
        if(ExpandedNodes[-1][0] == goal):
            break
        for item in tree[node]:
            if item[0] in [closed_item[0] for closed_item in ExpandedNodes]:
                break 
            
            cost.update({item[0]:cost[node]+item[1]}) 
            fn_node = cost[node] + heuristic[item[0]]+item[1] 
            Fringe.append([item[0],fn_node]) 


    deepNode = goal
    path = [goal]
    for i in range(len(ExpandedNodes)-2,-1,-1): 
        tempNode = ExpandedNodes[i][0]
        if deepNode in [children[0] for children in tree[tempNode]]:
            children_gn = [temp[1] for temp in tree[tempNode]]
            children_node = [temp[0] for temp in tree[tempNode]]

            if(cost[tempNode]+ children_gn[children_node.index(deepNode)]== cost[deepNode]): 
                path.append(tempNode)
                deepNode = tempNode
    path.reverse() 
    cost_of_goal = cost[goal]
    return path, cost_of_goal
```

لذلك بعد قراءة هذا البرنامج التعليمي ، يمكنك الحصول على لمحة عن سبب أهمية مشاكل الرسم البياني والشبكات بشكل عام في البرمجة نظرًا للمشكلة التي قدمناها ، فإن التنفيذ المثالي لخوارزمية الرسم البياني أو أي نوع من الخوارزمية يمكن أن يكون مهمًا في العالم الواقعي والأهم من ذلك أنه يمكن أن ينقذ العديد من الارواح.
